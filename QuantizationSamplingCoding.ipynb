```python
import matplotlib.pyplot as plt
import numpy as np

# plt.style.use('seaborn-poster')

# %matplotlib inline


fig, ax = plt.subplots(4, figsize= (16, 16))

f = 10
# sr = 100 # sampling rate
# ts = 1.0 / f # sampling interval
T = 1/f
t = np.arange(0, 2*T, T/100)
A = 10
y = A*np.sin(2*np.pi*f*t)
#figure 1
plt.tight_layout()
ax[0].plot(t,y)
ax[0].set_xlabel('Time(sec)')
ax[0].set_ylabel('Amplitude (volt)')
ax[0].set_title('Transmitting Singnals')

# sampling 
A = 500
Ts = T / 20
Fs = 1 / Ts
n = np.arange(0, len(y))
y1 = A*np.sin(2*np.pi*f*n*Ts)

#figure 2
plt.tight_layout()
ax[1].stem(n,y1)
ax[1].set_xlabel('Discrete time(sec)')
ax[1].set_ylabel('Amplitude (volt)')
ax[1].set_title('Discrete Time signal after sampling')
y2 = A + y1
#figure 3
plt.tight_layout()
ax[2].stem(n,y2)
ax[2].set_xlabel('Discrete time(sec)')
ax[2].set_ylabel('Amplitude (volt)')
ax[2].set_title('De level + Discrete time Signals')

# Quantization
y3 = np.round(y2)
#figure 4
plt.tight_layout()
ax[3].stem(n,y3)
ax[3].set_xlabel('Discrete time(sec)')
ax[3].set_ylabel('Amplitude (volt)')
ax[3].set_title('Quantization Signals')
plt.show()

y3 = np.array(y3, dtype= int)
n = np.max(y3)
n = np.ceil(np.log2(n))
n = int(n)
for e in y3:
    print(f'{e:0{n}b}')
```
