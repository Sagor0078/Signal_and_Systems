```python
import numpy as np
import matplotlib.pyplot as plt

x1=[4,3,6,1,4,9]
x2=[4,3,5,1,1,10]
n1=np.arange(-2,4)
n2=np.arange(-1,5) 
n=np.arange(min(min(n1),min(n2)),max(max(n1),max(n2))+1)
y1=np.zeros(len(n))
y2=np.zeros(len(n))

li=list(n)
y1[li.index(n1[0]):li.index(n1[0])+len(n1)]=x1
y2[li.index(n2[0]):li.index(n2[0])+len(n2)]=x2

# signal audition
y=y1+y2
plt.figure(figsize=(10,10))
plt.subplot(311)
plt.stem(n,y,basefmt="-b")
plt.title("Signal Audition")
plt.tight_layout()

# normal signal
x=[0,1,2,3,4,5,4,3];
n=np.arange(0,8)
plt.subplot(312)
plt.stem(n,x,basefmt="-b")
plt.title("Signal")
plt.tight_layout()

# folding signal
x.reverse()
n=np.flip(-n)
plt.subplot(313)
plt.stem(n,x,basefmt="-b")
plt.title("Folding Signal")
plt.tight_layout()
plt.show()
```
