```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0,51)
k = 2
plt.figure(figsize=(12,8))
plt.subplot(4,1,1)
plt.plot(t, k*np.abs(np.exp((-1/12)+1j*(np.pi/6)*t)))
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Absolute value  of  $x(t)=k*e^{-1/12 + j *pi/6* t}$');
plt.subplot(4,1,2) 
plt.plot(t, k*np.exp((-1/12)+1j*(np.pi/6)*t).real )
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Real part  of  $x(t)=k*e^{-1/12 + j *pi/6* t}$');
plt.subplot(4,1,3)
plt.plot(t, k*np.exp((-1/12)+1j*(np.pi/6)*t).imag )
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Imaginare part  of  $x(t)=k*e^{-1/12 + j *pi/6* t}$');
plt.subplot(4,1,4)
plt.plot(t, k*np.angle(np.exp((-1/12)+1j*(np.pi/6)*t))*360/(2*np.pi))
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Phase part  of  $x(t)=k*e^{-1/12 + j *pi/6* t}$');
plt.tight_layout()
```
