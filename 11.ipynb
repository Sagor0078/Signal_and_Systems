```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0,51)
k = 2
c = -(1/12) + (np.pi)*(0.16666666666j)  # 1/6 = 0.16666666666
y = k*np.exp(c*t)
# print(c,y)
plt.figure(figsize=(12,8))
plt.subplot(4,1,1)
plt.plot(t, np.abs(y))
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Absolute value  of  $x(t)=k*e^{c*t}$');
plt.subplot(4,1,2) 
plt.plot(t, y.real)
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Real part  of  $x(t)=k*e^{c*t}$');
plt.subplot(4,1,3)
plt.plot(t, y.imag )
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Imaginare part  of  $x(t)=k*e^{c*t}$');
plt.subplot(4,1,4)
plt.plot(t, np.angle((y)*360/(2*np.pi)))
plt.xlabel(r'$t$')
plt.ylabel(r'$|x(t)|$')
plt.title(r'Phase part  of  $x(t)=k*e^{c*t}$');
plt.tight_layout()
```
