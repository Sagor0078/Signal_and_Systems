```python
import numpy as np
import matplotlib.pyplot as plt

x1=[0,1,2,3,4]
x2=[2,2,2,2,2]
n1=np.arange(0,5)
n2=np.arange(-2,3) 
n=np.arange(min(min(n1),min(n2)),max(max(n1),max(n2))+1)
y1=np.zeros(len(n))
y2=np.zeros(len(n))

li=list(n)
y1[li.index(n1[0]):li.index(n1[0])+len(n1)]=x1
y2[li.index(n2[0]):li.index(n2[0])+len(n2)]=x2

fig , ax = plt.subplots(3,figsize = (16,16))
# Signal Multiplication
y=y1*y2
ax[0].stem(n,y)
ax[0].set_title('Signal Multiplication')
ax[0].set_xlabel('n')
ax[0].set_ylabel('x(n)')
# ax[0].set_tight_layout()
plt.tight_layout()

# signal shifting
#x(n)
n = np.arange(0,9)
x = [0,1,5,2,1,3,6,4,5]
ax[1].stem(n,x,'b', markerfmt= 'bo', basefmt= 'r')
ax[1].set_title('Signal shifting')
ax[1].set_xlabel('n')
ax[1].set_ylabel('x(n)')
plt.tight_layout()

#x(n-2) signal
ax[2].stem(n + 2, x, 'b', markerfmt= 'bo', basefmt= 'r')
ax[2].set_title('x(n-2) signal')
ax[2].set_xlabel('n')
ax[2].set_ylabel('x(n-2)')
plt.xticks(np.arange(len(n)+2))
plt.plot([0, 1, 2], [0, 0, 0], color= 'red')
plt.tight_layout()
plt.show()
```
